```python
import numpy as np 
import matplotlib.pyplot as plt

L = 100 # number of systems 
C = 1e-10 # capacitans
R = 10e11 # ohm
r = 1e6 # ohm
V0 = 0.1 # V

def Vpulse(t, t0, V0):
    """
    This function models the pulse
    as a sinusodial pulse
    """
    Vs = V0 * np.sin(np.pi * t / t0) * (0 <= t and t <= t0)
    
    return Vs
    
T = 0.1 # simulation in seconds
t0 = 0.01 # pulse length
dt = 0.1 * C * r
nsteps = int(T/dt)

t1 = np.linspace(0, T, nsteps)
Vs = np.zeros((nsteps, 1), float)

for i in range(nsteps):
    Vs[i] = Vpulse(t1[i], t0, V0)


plt.figure(figsize=(12,7))
plt.plot(t1, Vs)
```

### Task 3

```python
t = np.zeros((nsteps, 1), float)
V = np.zeros((nsteps, L), float)
V[0,:] = V0 # initial condition at t=0

for j in range(nsteps-1):
    t[j+1] = t[j] + dt
    V[j+1,0] = Vpulse(t[j+1], t0, V0) # left boundary condition i=0
    
    for i in range(1, L-1): # internal elements 
        V[j+1,i] = V[j,i] + (dt / C) * ((V[j,i+1] - 2 * V[j,i] + V[j,i-1]) / r - V[j,i] / R)

        
    # right boundary condition i=L-1
    V[j+1,L-1] = V[j,L-1] + (dt / C) *((-V[j,L-1] + V[j,L-2]) / r - V[j,L-1] / R)
```
