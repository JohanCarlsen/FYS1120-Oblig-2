```python
import numpy as np 
import matplotlib.pyplot as plt

L = 100 # number of systems 
C = 1e-10 # capacitans
R = 10e11 # ohm
r = 1e6 # ohm

def Vpulse(t, t0, V0):
    """
    This function models the pulse
    as a sinusodial pulse
    """
    Vs = V0 * np.sin(np.pi * t / t0) * (0 <= t and t <= t0)
    
    return Vs
    
T = 0.1 # simulation in seconds
t0 = 0.01 # pulse length
dt = 0.1 * C * r
nsteps = int(T/dt)

t1 = np.linspace(0, T, nsteps)
Vs = np.zeros((nsteps, 1), float)

for i in range(nsteps):
    Vs[i] = Vpulse(t1[i], t0, V0)


plt.figure(figsize=(12,7))
plt.title('Pulse shape')
plt.xlabel('time [s]')
plt.ylabel('Pulse strength [V]')
plt.plot(t1, Vs, label='Sinusodial pulse')
plt.legend(loc='right')
```

### Task 3

Koden under er hentet fra læreboken. Vi har kabellikningen $$C\frac{dV_i}{dt}+\frac{V_i}{R}=\frac{V_{i+1}-2V_i+V_{i-1}}{r}$$ som vi modifiserer til $$V_i(t+\Delta t)=V_i(t)-\frac{\Delta t}{RC}V_i(t)+\frac{\Delta t}{rC}\Big(V_{i+1}(t)-2V_i(t)+V_{i-1}(t)\Big)$$ Dette gir oss en funksjon $V(t, x)$, hvor $t$ er tiden og $x$ er posisjonen, eller hvilket system vi ser på.

```python
t = np.zeros((nsteps, 1), float)
V = np.zeros((nsteps, L), float)
V[0,:] = 0 # initial condition at t=0

for j in range(nsteps-1):
    t[j+1] = t[j] + dt
    V[j+1,0] = Vpulse(t[j+1], t0, V0) # left boundary condition i=0
    
    for i in range(1, L-1): # internal elements 
        V[j+1,i] = V[j,i] + (dt / C) * ((V[j,i+1] - 2 * V[j,i] + V[j,i-1]) / r - V[j,i] / R)
        
    # right boundary condition i=L-1
    V[j+1,L-1] = V[j,L-1] + (dt / C) *((-V[j,L-1] + V[j,L-2]) / r - V[j,L-1] / R)
```

```python
plt.figure(figsize=(12,7))
plt.title('Pulse signal for 10 systems over time')
plt.xlabel('time [t]')
plt.ylabel('Pulse strength [V]')

for i in range(0, L-1, 10):
    plt.plot(t, V[:,i])
```

# * something about the plot *

```python
plt.figure(figsize=(12,7))
plt.title('Pulse signal for 10 systems over time, normalized')
plt.xlabel('time [t]')
plt.ylabel('Pulse strength [V]')

for i in range(0, L-1, 10):
    plt.plot(t, V[:,i] / max(V[:,i]))
```

# * something about the plot *

```python
plt.figure(figsize=(12,7))
time = 1000 * dt
plt.title(f'Pulse signal at all systems at $t\in[{time:.2f}, {nsteps-1:.2f}]$ s, normalized')
plt.xlabel('System number')
plt.ylabel('Pulse strength [V]')

for j in range(1000, nsteps-1, 1000):
    plt.plot(V[j,:] / max(V[j,:]))
```

### Task 4

Jeg setter $\text{X}$-komponenten til å være et batteri $e$ koblet i serie med resistansen $R$. Det gir $$I_{x,j}=\frac{V_0+e}{R}$$ Jeg endrer også $R$ litt, for å se tydeligere hva som skjer med spenningen.

```python
t_new = np.zeros((nsteps, 1), float)
V_new = np.zeros((nsteps, L), float)
V_new[0,:] = 0 # initial condition at t=0
e = 20e-3 # mV
R_new = 1e11 # ohm

for j in range(nsteps-1):
    t_new[j+1] = t_new[j] + dt
    V_new[j+1,0] = Vpulse(t_new[j+1], t0, V0) # left boundary condition i=0
    
    for i in range(1, L-1): # internal elements 
        V_new[j+1,i] = V_new[j,i] + (dt / C) * ((V_new[j,i+1] - 2 * V_new[j,i] + V_new[j,i-1]) / r - (V_new[j,i] + e) / R_new)

        
    # right boundary condition i=L-1
    V_new[j+1,L-1] = V_new[j,L-1] + (dt / C) *((-V_new[j,L-1] + V_new[j,L-2]) / r - (V_new[j,L-1] + e) / R_new)
```

```python
plt.figure(figsize=(12,12))
plt.subplot(211)
plt.title('Pulse signal for 10 systems over time, normalized')
plt.ylabel('Pulse strength [V]')

for i in range(0, L-1, 10):
    plt.plot(t_new, V_new[:,i] / max(V_new[:,i]))
    
plt.subplot(212)
plt.title('Same plot as earlier, for reference')
plt.xlabel('time [t]')
plt.ylabel('Pulse strength [V]')

for i in range(0, L-1, 10):
    plt.plot(t, V[:,i] / max(V[:,i]))
```

# * something about the plots *

```python
plt.figure(figsize=(12,12))
plt.subplot(211)
time = 1000 * dt
plt.title(f'Pulse signal at all systems at $t\in[{time:.2f}, {nsteps-1:.2f}]$ s, normalized')
plt.ylabel('Pulse strength [V]')

for j in range(1000, nsteps-1, 1000):
    plt.plot(V_new[j,:] / max(V_new[j,:]))
    
plt.subplot(212)
plt.title(f'Same plot as earlier, for reference')
plt.xlabel('System number')
plt.ylabel('Pulse strength [V]')

for j in range(1000, nsteps-1, 1000):
    plt.plot(V[j,:] / max(V[j,:]))
```
